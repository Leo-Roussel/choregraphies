## E.1 Le problème à $N$-corps
On souhaite trouver des solutions périodiques du problèmes à $N$-corps dans le plan. Nous nous plaçons dans le cas particulier d'un système de $N$ corps ponctuels de même masse. On suppose que ce système est isolé du reste de l'univers et qu'il évolue  uniquement sous l'action des forces de gravitation que les masses ponctuelles exercent les unes sur les autres.

Notons $q^0(t),\dots,q^{N-1}(t)$ les positions des $N$-masses à l'instant $t$. La position du corps $j$ à l'instant $t$ est représentée par le vecteur $q^j(t)\in\mathbb{R}$
$$
q^j(t)=\binom{x^j(t)}{y^j(t)},\quad\text{pour }j=0,\dots,N-1.
$$
L'équation qui régit l'évolution du système est
$$
\ddot q^j=-Gm\sum_{k\neq j} \dfrac{q^k-q^j}{|q^k-q^j|^3},\quad\text{pour }j=0,\dots,N-1, 
$$
où $G$ est la contante de gravitation universelle et $m$ la masse commune des masse ponctuelles. _(On note $|x|$ la norme euclidienne d'un vecteur $x$ de $\mathbb{R}^2$)._

À partir de maintenant, pour $j\in\mathbb{Z}$,  $q^j$ désigne $q^{\{j\}}$ où $\{j\}$  est le reste de la division euclidienne de $j$ par $N$ $(j=Nq+\{j\}$ avec $0\le j<N$). Par exemple on peut réécrire la formule ci-dessus :
$$
\ddot q^j=-Gm\sum_{k=1}^{N-1} \dfrac{q^{j+k}-q^j}{|q^{j+k}-q^j|^3},\quad\text{pour }j=0,\dots,N-1, 
$$

En faisant le changement d'unité de distance $q^j=\lambda \breve{q}^j$ pour un $\lambda>0$, on a 

$$
\lambda  \ddot{\breve{q}}^j=-\dfrac{Gm}{\lambda^2}\sum_{k\neq j} \dfrac{\breve{q}^k-\breve{q}^j}{|\breve{q}^k-\breve{q}^j|^3}.\quad\text{pour }j=0,\dots,N-1
$$
En choisissant $\lambda:=(Gm)^{1/3}$, on se ramène à 

$$
\ddot {\breve{q}}^j=-\sum_{k\neq j} \dfrac{\breve{q}^k-\breve{q}^j}{|\breve{q}^k-\breve{q}^j|^3},\quad\text{pour }j=0,\dots,N-1.
$$
De la sorte, on peut toujours supposer $Gm=1$. Nous allons donc plutôt étudier le système 
$$
\tag{1}
\ddot q^j=-\sum_{k=1}^{N-1} \dfrac{q^{j+k}-q^j}{|q^{j+k}-q^j|^3},\quad\text{pour }j=0,\dots,N-1.
$$

## E.2 Solutions périodiques 
On souhaite trouver des solutions périodiques de (1), c'est à dire qu'on cherche des solutions telles qu'il existe un temps $T_{\text{per}}>0$ avec 
$$
q^j(t+T_{\text{per}})=q^j(t) \quad\text{pour }t\in\mathbb{R}\text{ et }j=0,\dots,N-1.
$$

Nous proposons un nouveau changement de variable pour se ramener à $T_{\text{per}}=NT$ où $T>0$ est fixé arbitrairement. On pose $z^j(s)=\mu q^j((T_{\text{per}}s/(NT))$ pour un $\mu>0$ positif, de sorte que les $z^j$ sont $NT$-périodiques. Les équations (1) sont équivalentes à

$$
\dfrac1\mu\left(\dfrac{NT}{T_{\text{per}}}\right)^2\ddot z^j=-\mu^2\sum_{j\neq i} \dfrac{z^k-z^j}{|z^k-z^j|^3},\quad\text{pour }j=0,\dots,N-1.
$$

En prenant, $\mu=(NT/T_{\text{per}})^{2/3}$, on se ramène à chercher une application $t\in\mathbb{R} \mapsto(z^0,\dots,z^{N-1})(t)\in\left(\mathbb{R}^2\right)^N$,  qui soit $NT$-périodique et solution de (1). 

On peut donc se contenter d'étudier les solutions $NT$-périodiques de (1) pour un $T>0$ arbitraire.

## E.3 Solutions triviales

Il existe des solutions particulières dites solutions triviales à ce problème : celle qui sont formées par $N$ corps placées de manière équidistantes sur un cercle et qui se pourchassent à vitesse constante sur ce cercle. Supposons le cercle centré en 0 et notons $R$ son rayon. 

À rotation (ou translation en temps près) on peut supposer $q^0(0)=(R,0)$. Quitte à renuméroter les autres masses, on peut aussi supposer 
$$
q^j(0)=R\,\left(\cos (j\omega_N),\sin (j\omega_N)\right)\,\quad\text{pour }j=0,\dots,N-1,
$$
avec $\omega_N= 2\pi/N$.

Notons $\Omega$, la vitesse angulaire de rotation du système. Si la plus petite période du système est $NT$, on a : $\Omega=\pm 2\pi/(NT)$. La position des masses au cours du temps est décrite par

$$
\tag{2} q^j(t)=R\, \left(\cos (j\omega_N +\Omega t),\sin (j\omega_N +\Omega t)\right),\quad\text{pour }t\in\mathbb{R}\text{ et }j=0,\dots,N-1,
$$

```python
### SOLUTION 11.c
def calcule_R(N):
    S_N=sum(1/np.sin(np.linspace(1,N-1,N-1)*np.pi/N))
    return  (S_N*N**2/(2*np.pi**2))**(1/3)/2 

def calcule_Req(N):
    return (np.log(N))**(1/3)*N/(2*np.pi)
```

```python
### Verification de la limite de R(N)/N

R,N,Req = [],[],[]

for i in range(1,21):
    n=2**i
    r =calcule_R(n)
    req=calcule_Req(n)
    R.append(r) ; Req.append(req) ; N.append(n)

R=np.asarray(R) ; Req=np.asarray(Req) 

plt.semilogx(N,np.log(R/Req - 1),'.')

plt.title(r"graphe de $R(N)/R_{eq}(N) -1$")

plt.show()
```

Pour visualiser le mouvement du système de masses ponctuelles, nous pouvons utiliser la bibliothèque `matplotlib.animation`.

Commençons par discrétiser l'intervalle de temps $[0,NT]$. On se donne un entier $m\ge1$ et on pose $M=mN$ puis
$$
h=\dfrac Tm=\dfrac{NT}M.
$$
Pour $l\in\mathbb{Z}$, on note 
$$
t_l=jh,
$$
de sorte que la suite :
$$
t_0=0<t_1=h<\cdots<t_j<t_{j+1}<\cdots<t_{M-1}<t_M=NT
$$
forme une discrétisation de la période $[0,NT]$.

La position de $q^j$ au temps $t_l$ est donnée par 

$$
q^j_l:=\binom{R\cos(j\omega_N + \Omega t_l)}{R\sin(j\omega_N + \Omega t_l)}.  
$$
avec $\omega_N=2\pi/N$ et $\Omega=2\pi/(NT)$.

On remarque que comme $q^j(t)=q^0(t+jT)$ et que $t_l+jT = t_l +jmh=t_{l+jm}$. En notant $[k]$ le reste de la division euclidienne de $k$ par $M$ on a 
$$
q^j_l= q^0_{l+{jm}} = \binom{R\cos(j\omega_N + \Omega t_{[l+jm]})}{R\sin(j\omega_N + \Omega t_{[l+jm]})}\tag{*}
$$
Pour cette raison, nous allons nous contenter de discrétiser $q^0$.

```python
## paramètres du système
N, T = 5, 1
omega_N = 2*np.pi/N
Omega = omega_N/T
R = calcule_R(N)
##  paramètres de discrétisation
m=30
h=T/m
M=N*m

## temps discrets
tps = np.linspace(0,N*T-h,M)
## q0
q0 = np.asarray([R*np.cos(Omega*tps), R*np.sin(Omega*tps)])
print(f"forme de q0 : {np.shape(q0)}")
x, y = q0[0,:], q0[1,:]
```

On peut déjà tracer la trajectoire de $q^0$ sur une période :

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()
```

Noter que la trajectoire n'est pas fermée. Pour cela, il faut ajouter le premier point en fin de liste.

```python
X,Y=list(x), list(y)
X.append(x[0]) ; Y.append(y[0])

plt.figure(figsize=(6,6))

# trajectoire de q^0
plt.plot(X,Y)

plt.xlabel(r"$x$", fontsize=20)
plt.ylabel(r"$y$", fontsize=20)
plt.axis("equal")
plt.title(r"trajectoire de $q^0$",fontsize=20, pad=20)
plt.show()
```

Nous voulons aussi représenter les $N$ masses à un instant fixé sur cette courbe. Pour cela nous utilisons l'identité (*) plus haut : les coordonnées de la masse $j$ au temps $t_l$ seront $(x_{[l+jm]},y_{[l+jm]})$ _(où $[k]$ désigne le reste de la division euclidienne de l'entier $k$ par $M$)_. 

Pour obtenir $[k]$ avec python, nous utilisons l'opération `%`. En effet, l'instruction  `k%M` renvoie le reste de la division euclidienne de l'entier $k$ par $M$.

```python
# Exemples
for a,b in ((7,3),(12,5), (107,50)):
    print(f"Reste de la division de {a} par {b} : {a}%{b} = {a%b}")
```

Dans la figure suivante nous représentons la trajectoire et les positions des masses à l'intant discret précédent immédiatement $t=0,27$.

```python
plt.figure(figsize=(6,6))
# trajectoire de q^0
plt.plot(X,Y) 

# position des N masse au temps t_l tel que t-h <t_l≤ t
t = 0.27
l = int(m*t)
for j in range(N):      
    plt.plot(x[(l + j*m)%M],y[(l + j*m)%M],'o')

plt.xlabel(r"$x$", fontsize=20)
plt.ylabel(r"$y$", fontsize=20)
plt.axis("equal")
titre =f"Position des masses à l'intant t={t}, dans le cas m={m}, on a l={l}"
plt.title(titre,fontsize=20, pad=20)
plt.show()
```

Nous voulons maintenant voir les masses en mouvement. Nous importons la bibliothèque `animation` de `matplotlib` pour pouvoir créér un film montrant l'évolution du système de masses.

```python
from matplotlib import animation
```

Nous allons maintenant utilser `animation` pour représenter l'évolution du système dans un petit film de $m$ images qui se répète de maniere périodique.

```python
%matplotlib notebook

xmin, xmax = 1.1*np.min(x), 1.1*np.max(x)
ymin, ymax = 1.1*np.min(y), 1.1*np.max(y)

fig = plt.figure(figsize=(8,8)) # initialise la figure et les caractéristique du plot

plt.xlim(xmin, xmax) ; plt.ylim(ymin,ymax)
plt.xlabel(r"$x$",fontsize=20) ; plt.ylabel(r"$y$",fontsize=20)
plt.title(f"système de {N} masses en rotation",fontsize=30,pad=20)
plt.axis("equal")

X, Y =list(x), list(y)
X.append(x[0]) ; Y.append(y[0])
line = plt.plot(X,Y,'-b',[], [],'go')

# fonction à définir quand blit=True
# crée l'arrière de l'animation qui sera présent sur chaque image
def init():
    line[1].set_data([],[])
    return line,

def animate(i): 
    xs, ys = [], []
    for j in range(N + 1):
        xs.append(x[(i + j*m)%M])
        ys.append(y[(i + j*m)%M])
    line[1].set_data(xs,ys)
    return line,
    
nbframes=m
intervalle =int(1200/m)
anim=animation.FuncAnimation(fig, animate, init_func=init, frames=nbframes, 
                               blit=True, interval=intervalle, repeat=True)
plt.show()
```

Les solution triviales (2) sont des solutions périodiques très particulières du problème à $N$ corps. En particulier, pour ces solutions, tous les corps parcourent la _même_ trajectoire avec un déphasage les uns par rapport aux autres. 

Plus précisément 
$$
q^{j+k}(t)=q^j(t+kT)\quad \text{pour }t\in\mathbb{R},\ j=0,\dots,N-1\text{ et }k\in
\mathbb{Z}.
$$

_(Nous utilisons encore la convention $q^{j+k}=q^{\{j+k\}}$)._

En notant $\tau$ l'opérateur de translation en temps défini par $\tau x(t):=x(t+T)$, l'identité précédente se réécrit 

$$\tag{3}
q^j=\tau^jq^0\quad \text{pour } j\in\mathbb{Z}.
$$

Nous appelons _chorégraphies_ les solutions de (1) qui satisfont (3). 

<a Id="exo12"></a>__Exercice 12__

Montrer que (3) entraîne que $t\mapsto (q^0,\dots,q^{N-1})$ est $NT$-périodique.

<div class="alert alert-block alert-info" style="margin-top: 0px">

On pose $f : t \mapsto (q^0,\,...\,,q^{N-1})$ ; calculons $f(t+NT)$. Pour $j=0,\dots,N-1$, on a :
$$ q^j(t+NT) = q^{j+N}(t) \text{.} $$

Or comme on définit, pour tout $k \in \mathbb{Z}, q^{k} = q^{\{k\} }$, on obtient l'expression, pour $j=0,\dots,N-1$,
$$ q^{j+N}=q^j \text{,}$$

d'où la $NT$-périodicité de $f$.

## E.6 Énergies potentielle et cinétique

Les forces de gravitations sont conservatives. C'est à dire qu'elles sont associées à une énergie potentielle. Deux corps ponctuels de masses $m_1$, $m_2$ à une distance $d$, l'un de l'autre ont une énergie potentielle
$
\dfrac{-Gm_1m_2}d.
$

Cete énergie est négative et croît vers 0 quand la distance $d$ tends vers $+\infty$. Au contraire elle tends vers $-\infty$ quand les corps deviennent infiniment proches. 

Avec notre normalisation, nous pouvons prendre $G=1$ et toutes les masses égales à 1. L'énergie potentielle associée aux $N$ masses de positions  $(q^1,\dots,q^N)$ est 
$$
\mathcal{U}(q^1,\dots,q^N):=-\dfrac12\sum_{j=0}^{N-1}\sum_{k=1}^{N-1}\dfrac1{|q^{j+k}-q^j|}.
$$
Le facteur $1/2$ vient du fait que chaque paire de corps apparaît exactement deux fois dans la somme.

À un corps ponctuel de masse $m$ et de vitesse $v\in\mathbb{R}^2$ est associée l'énergie cinétique 
$
\dfrac12 m|v|^2.
$
Ici l'énergie cinétique associée à $N$ corps de masse 1 et de vitesses $(\dot q^1,\dots,\dot q^N)$ est 

$$
\mathcal{E}_c(\dot q^1,\dots,\dot q^N):=\dfrac12\sum_{j=0}^{N-1} \left|\dot q^j\right|^2.
$$

L'énergie totale du système (son Hamiltonien) est
$\quad
\mathcal{H}(q,\dot q):=\mathcal{E}_c(\dot q) + \mathcal{U}(q).
$

On définit aussi son Lagrangien, 
$\quad
\mathcal{L}(q,\dot q):=\mathcal{E}(\dot q) - \mathcal{U}(q).
$

$$\forall j \in \mathbb{Z}, \forall t \in [0,NT], q^j(t) = q^0(t+jT) = \tau^jq^0(t)\text{.} $$ 
    
On remplace dans l'expression de $A[q]$ :
$$ A[q] = \int_0^T \left[\frac{1}{2}\sum_{j=0}^{N-1}|\tau^j\dot{q}^0|^2 + \frac{1}{2}\sum_{j=0}^{N-1}\sum_{k=1}^{N-1}\frac{1}{|\tau^jq^0 - \tau^k\tau^jq^0|}\right]\,dt = \frac{1}{2}\int_0^T \sum_{j=0}^{N-1}|\tau^j\dot{q}^0|^2\,dt + \frac{1}{2}\int_0^T\sum_{j=0}^{N-1}\sum_{k=1}^{N-1}\frac{1}{|\tau^jq^0 - \tau^k\tau^jq^0|}\,dt \text{.}$$
    
On effectue alors le changement de variable suivant : soit $\varphi_j : t \mapsto s = t+jT$, $\varphi_j$ est un $C^1$-difféomorphisme et sa dérivée est identiquement égale à 1 donc on peut écrire d'une part, pour le terme de gauche,
$$ \frac{1}{2}\int_0^T\sum_{j=0}^{N-1}|\tau^j\dot{q}^0|^2\,dt = \frac{1}{2}\sum_{j=0}^{N-1}\int_{jT}^{(j+1)T} |\dot{q}^0|^2\,dt = \frac{1}{2}\int_0^{NT}|\dot{q}^0|^2\,dt\text{,}$$

et d'autre part pour le terme de droite :

$$ \boxed{A[q] = \int_{0}^{NT} \left[\frac{1}{2}|\dot{q}^0|^2 + \frac{1}{2}\sum_{k=1}^{N-1} \frac{1}{|q^0 - \tau^kq^0|}\right]\,dt \text{.}}$$

On fixe $T>0$ et on s'intéresse aux chorégraphies à $N$ corps de période $NT$. Les corps ponctuels ont tous la même masse $m=1$. 

Comme dans la partie **E.4.**, on discrétise l'intervalle de temps $[0,NT]$. On se donne un entier $m\ge1$ et on pose $M=mN$ puis $h=T/m=NT/M$.
Pour $l\in\mathbb{Z}$, on note $t_l=lh$,
et la suite :
$$
t_0=0<t_1=h<\cdots<t_j<t_{j+1}<\cdots<t_{M-1}<t_M=NT
$$
forme une discrétisation de la période $[0,NT]$.

Pour $q^0=\binom{x^0}{y^0}$, on va utiliser deux vecteurs $x$, $y$ de longueur $M$, tels que 
$$
x^0_l=x(t_j),\quad y^0_l=y(t_l)\quad\text{pour }l\in\{0,\dots M-1\}.
$$
On a alors (en utilisant la périodicité)
$$
x^k(t_l)=\tau^k x^0(t_l+kT)=x^0(t_{[l+km]}) = x_{[l+km]}.
$$
et de même
$$
y^k(t_l)=y_{[l+km]}.
$$

### Implémentation

Les vecteurs $x^0$ et $y^0$ vont être représentés par deux tableaux numpy $x$, $y$ de longueur $M$. Pour le beoisn des exemples ci-dessous, commençons par construire une paire $(x,y)$ représentant une discrétisation d'une trajectoire circulaire.

```python
T=1
N=5
m=20
h=T/m
M=m*N
NT=N*T
tps = np.linspace(0,N*T - h,M)
x = np.cos(2*np.pi*tps/NT)
y = np.sin(2*np.pi*tps/NT)
```

### Discrétisation de l'énergie

Pour discrétiser l'énergie cinétique, on utilise le taux d'accroissement $\dfrac{q^0(t_{l+1})-q^0_l}h$ pour approcher la dérivée $\dot q^0(t_l)$ On a alors 
$$
\mathcal{E}_c^h[q^0]=\dfrac12\left|\dfrac{q^0(t_{l+1})-q^0_l}h\right|^2 =\dfrac1{2h^2}\sum_{l=0}^{M-1}\left[(x_{[l+1]}- x_l)^2+(y_{[l+1]}- y_l)^2\right]
$$

Pour discrétiser l'énergie potentielle au temps $t_l$, nous n'avons pas d'approximation à faire. On pose :
$$
\begin{array}{rcl}
\mathcal{U}^h[q^0](t_l)&=&
\dfrac12\displaystyle
\sum_{k=1}^{N-1} \dfrac1{\left|q^0(t_l)-\tau^k q^0(t_l)\right|}\\
&=&\dfrac12
\displaystyle\sum_{k=1}^{N-1} \dfrac1{\left|q^0(t_l)- q^0(t_{[l+km]})\right|}\\
&=&\dfrac12\displaystyle\sum_{l=0}^{M-1}h\sum_{k=1}^{N-1} \left[(x_{[l+km]}- x_l)^2+(y_{[l+km]}- y_l)^2 \right]^{-1/2}.
\end{array}
$$

Pour l'intégration en temps, on utilise la formule des rectangles *(comme on l'applique à des fonctions périodiques, cette méthode est d'ordre infini)*. La discrétisation de $F[q^0]$ est donc 
$$
F^h(x,y):=\dfrac1{2h}\sum_{l=0}^{M-1}\left[(x_{[l+1]}- x_l)^2+(y_{[l+1]}- y_l)^2\right]+\dfrac h2\sum_{l=0}^{M-1}\sum_{k=1}^{N-1} \left[(x_{[l+km]}- x_l)^2+(y_{[l+km]}- y_l)^2 \right]^{-1/2}.
$$

Faites varier $m$ dans l'ensemble $\{30,60,120,240,480\}$ ? Comment varie le nombre d'itérations nécessaire à la convergence ?

<div class="alert alert-block alert-info" style="margin-top: 0px">

<b>Solution 16.c</b>
    
Le nombre d'itérations nécessaire à la convergence varie de manière équivalente au nombre de points de discrétisation : si $m = 30 \times 2^k$, $k=0,\dots,4$ alors $i_{req} = 3 \times 2^k$.

```python
T, N = 1, 5
m=30
h, M, NT =T/m, m*N, N*T

tps = np.linspace(0,N*T - h,M)
x = np.cos(2*np.pi*tps/NT)
y = np.sin(2*np.pi*tps/NT)

base = np.asarray(range(M))
plus, moins = (base + 1)%M, (base - 1)%M

nitermax = 4000
imax = 10
niter = 0
f = energie(x,y,h,N,M,m,plus,moins,base)
g_x,g_y = gradient(x,y,h,N,M,m,plus,moins,base)
K = np.concatenate((g_x,g_y))
H = np.dot(K,K)
norm_g = np.sqrt(H)
norm_g0 = norm_g
c1, beta, tol = 0.5, 0.75, 10**(-5)
t = h/2

F, I, G = [f],[],[norm_g]


# boucle d'optimisation
while norm_g>tol*norm_g0 and niter<nitermax:   
    # line search 
    i, t = 0, 1.2*t 
    xt,yt = x - t*g_x, y - t*g_y
    ft = energie(xt,yt,h,N,M,m,plus,moins,base)
    while ft>=f - c1*t*H and i<imax:  
        t = beta*t
        xt,yt = x - t*g_x, y - t*g_y
        ft = energie(xt,yt,h,N,M,m,plus,moins,base)
        i+=1
    
    # updates
    x, y, f = xt, yt, ft
    g_x, g_y = gradient(x,y,h,N,M,m,plus,moins,base)
    K = np.concatenate((g_x,g_y))
    H = np.dot(K,K)
    norm_g = np.sqrt(H)
    niter+=1   
    
    # stockages
    I.append(i)
    G.append(norm_g/norm_g0)
    F.append(f)
```

```python
def descent(x,y,T,N,m,t,tol):
    h, M, NT =T/m, m*N, N*T
    base = np.asarray(range(M))
    plus, moins = (base + 1)%M, (base - 1)%M
    niter, nitermax, imax = 0, 4000, 10
    f = energie(x,y,h,N,M,m,plus,moins,base)
    g_x,g_y = gradient(x,y,h,N,M,m,plus,moins,base)
    K = np.concatenate((g_x,g_y))
    norm_g = np.sqrt(np.dot(K,K))
    norm_g0 = norm_g
    while norm_g>tol*norm_g0 and niter<nitermax:   
        i, t = 0, 1.2*t 
        xt,yt = x - t*g_x, y - t*g_y
        ft = energie(xt,yt,h,N,M,m,plus,moins,base)
        while ft>=f - c1*t*np.dot(K,K) and i<imax:  
            t = beta*t
            xt,yt = x - t*g_x, y - t*g_y
            ft = energie(xt,yt,h,N,M,m,plus,moins,base)
            i+=1
        x, y, f = xt, yt, ft
        g_x, g_y = gradient(x,y,h,N,M,m,plus,moins,base)
        K = np.concatenate((g_x,g_y))
        norm_g = np.sqrt(np.dot(K,K))
        niter+=1 
    return x, y
```

Nous reprenons ici la méthode de visualisation de la partie **E.4** mais en créant cette fois un film qui peut être relancé à tout moment. 

Cette méthode peut nécessiter l'installation de `ffmpeg` s'il n'est pas déjà présent dans votre système. Si vous utilisez anaconda : faire `conda install ffmpeg` dans un terminal ou bien utilisez le navigateur d'Anaconda pour demander l'installation  de `ffmpeg` en suivant l'onglet `Environments`.

On obtient ensuite la visualisation par la commande `HTML(anim.to_html5_video())` (qui prend du temps). On peut relancer la vidéo à volonté en cliquant dessus (double cliquer pour le mode plein écran).

```python
## paramètres du système
N, T = 5, 1
omega_N = 2*np.pi/N
Omega = omega_N/T
R = calcule_R(N)
##  paramètres de discrétisation
m=50
h=T/m
M=N*m

## temps discrets
tps = np.linspace(0,N*T,M)
## q0
q0 = np.asarray([R*np.cos(Omega*tps), R*np.sin(Omega*tps)])
print(f"forme de q0 : {np.shape(q0)}")
x, y = q0[0,:], q0[1,:]
```

```python
from matplotlib import animation
from IPython.display import HTML

def film(x,y,N,m,M):    
    %matplotlib notebook
    xmin, xmax = 1.1*np.min(x), 1.1*np.max(x)
    ymin, ymax = 1.1*np.min(y), 1.1*np.max(y)

    fig = plt.figure() # initialise la figure et les caractéristique du plot
    plt.xlim(xmin, xmax) ; plt.ylim(ymin,ymax)
    plt.xlabel("x") ; plt.ylabel("y")

    plt.title(f"animation : système de {N} masses en intéractions")
    plt.axis("equal")
    #fig.set_size_inches(6, 6)
    X, Y =list(x), list(y)
    X.append(x[0]) ; Y.append(y[0])
    line = plt.plot(X,Y,'-b',[], [],'go')

    # fonction à définir quand blit=True
    # crée l'arrière de l'animation qui sera présent sur chaque image
    def init():
        line[1].set_data([],[])
        return line,

    def animate(i): 
        xs, ys = [], []
        for j in range(N + 1):
            xs.append(x[(i + j*m)%M])
            ys.append(y[(i + j*m)%M])
        line[1].set_data(xs,ys)
        return line,

    nbframes=m
    intervalle =int(1200/m)
    anim=animation.FuncAnimation(fig, animate, init_func=init, frames=nbframes, 
                                   blit=True, interval=intervalle, repeat=True)
    plt.close()
    %matplotlib inline
    return anim
```

```python
anim = film(x,y,N,m,M)
```
